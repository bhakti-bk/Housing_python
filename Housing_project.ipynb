```python
"""Q3.Show with the help of visualization, how median_income and median_house_values are related?"""

#we have two variable which are continuous in nature. we have achieved this with 3 approaches. 
#first is with simple plotting method and then with scatter and regression plot. 
```

```python
#first approach is plotting method in matplotlib

x = [housing.median_income]                                    # defining x axis
y = [housing.median_house_value]                               #defining y axis
plt.plot(x, y, marker= 'o', mec='b', mfc='w', ms=7)            # plotting x and y axis, size and adding color
plt.xlabel('median_income')                                      #giving title to x axis
plt.ylabel('median_house_value')                                   #giving title to y axis
plt.title('Relation between median income and median house value')  # giving title to the figure
plt.show()
```

```python
#second approach is scatter plot in seaborn lib

ax= sns.scatterplot(x= 'median_income', y= 'median_house_value',hue= 'median_income', data=housing, marker= 'o', facecolor='c', s=25)
plt.show()                  #plotting scatter plot
```

```python
#third approach is regression plot as both are quantitative(continuous) data

sns.set_style('whitegrid')
sns.lmplot(x= 'median_income', y= 'median_house_value',data = housing,fit_reg= True ,line_kws={'color':'red'}) 
plt.show()             #to plot regression we use fit_regression and to change color of that line used line_kws
    
```

```python
"""Data is continuous. median_house_value increases as median_income increases."""
```

```python
"""10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large."""

housing.loc[housing['total_bedrooms']<= 10, 'total_bedroom_size']= 'small' #use loc for slicing and define conditions 
housing.loc[housing['total_bedrooms']>10, 'total_bedroom_size'] = 'medium' 
housing.loc[housing['total_bedrooms']<1000 , 'total_bedroom_size'] = 'medium'
housing.loc[housing['total_bedrooms']>=1000,  'total_bedroom_size']= 'large'
```
